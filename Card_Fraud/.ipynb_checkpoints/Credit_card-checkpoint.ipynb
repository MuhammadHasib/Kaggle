```python
data = data.sample(frac = 1, random_state = 1)
print(data.shape)
```

```python
# Plot histogram of each parameter
data.hist(figsize = (20,20))
plt.show()
```

```python
# Determine number of froud case in dataset
Fraud = data[data['Class']==1]
Valid = data[data['Class']==0]

outlier_fraction = len(Fraud) / float(len(Valid))
print(outlier_fraction)

print('Fraud Class: {}'.format(len(Fraud)))
print('Valid Class: {}'.format(len(Valid)))
```

```python
# correlation matrix
corrmat = data.corr()
fig = plt.figure(figsize = (12,9))

sns.heatmap(corrmat, vmax = .8, square = True)
plt.show()
```

```python
# Gel all the column from the DataFrame
columns = data.columns.tolist()

# Filter the columns to remove data we do not want
columns = [c for c in columns if c not in ['Class']]

# store the variab le we'll be predictiong on

target = 'Class'

X = data[columns]
Y = data[target]

# Print the shapes of X and Y
print(X.shape)
print(Y.shape)
```

```python
from sklearn.metrics import classification_report, accuracy_score
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor

# define the radom state
state = 1

#define the outlier detection method
classifers = {
    
    "Isolation Forest": IsolationForest(max_samples=len(X), contamination = outlier_fraction,random_state= state),
    "Local Outlier Factor": LocalOutlierFactor(n_neighbors = 20, contamination = outlier_fraction)
    
}
```

```python
# Fit the model
n_outliers = len(Fraud)

for i, (clf_name, clf) in enumerate(classifers.items()):
    
    # fir the data and tag outliersabs
    if clf_name == "Local Outlier Factor":
        y_pred = clf.fit_predict(X)
        scores_pred = clf.negative_outlier_factor_
    else:
        clf.fit(X)
        scores_pred = clf.decision_function(X)
        y_pred = clf.predict(X)
        
        # Reshape the prediction values to 0 fr Valid, 1 for Fraud
        
        y_pred[y_pred == 1] = 0
        y_pred[y_pred == -1] = 1
        
        n_errors = (y_pred != Y).sum()
        
        # Run the classifcation metrics
        print('{}:{}'.format(clf_name, n_errors))
        print(accuracy_score(Y, y_pred))
        print(classification_report(Y,y_pred))
```
